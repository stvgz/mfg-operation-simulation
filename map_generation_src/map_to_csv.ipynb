```python
# define functon to add style into json
def add_style(j):
    j['properties']['style'] = {
        "__comment": "all SVG styles allowed",
        "color":"red",
        "fill":"red",
        "stroke-width":"3",
        "fill-opacity":0.6
    }
    
    return j
```

```python
# apply d1
d1['styled'] = d1['formated'].apply(lambda x: json.dumps(add_style(json.loads(x))))
```
