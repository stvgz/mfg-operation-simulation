```python
# Generate random numbers representing consumption of each part each month
import random

for part in spareparts_dict:
  spareparts_dict[part] = random.randint(0, 100)
  
print(spareparts_dict)
```

```python
# Create a table of each month start from 2020.01 to 2023.12, Generate random numbers representing consumption of each spart parts saved in spareparts_dict with python code
```

```python
import pandas as pd
import random


# Create a list of months from 2020/01 to 2023/12
months = [f'2020/{i:02d}/01' for i in range(1, 13)] + \
         [f'2021/{i:02d}/01' for i in range(1, 13)] + \
         [f'2022/{i:02d}/01' for i in range(1, 13)] + \
         [f'2023/{i:02d}/01' for i in range(1, 13)]

# Generate random numbers representing consumption of each spart parts
for part in spareparts_dict.keys():
    
    spareparts_dict[part] = [random.randint(0, 10) for _ in range(len(months))]
    
    # generate random numbers following gaussian dist with mean of 10 and std of 5
    numbers = np.random.normal(10, 5, len(months))
    numbers = [int(i) if i>=0 else 0 for i in numbers]
    
    spareparts_dict[part] = [int(i) for i in numbers]

# Create a table with the data
df = pd.DataFrame(spareparts_dict, index=months)
print(df)
```
