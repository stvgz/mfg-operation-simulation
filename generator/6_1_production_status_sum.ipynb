# 计算每小时的状态

```python
# TODO: 检查一下数据是否有问题
```

```python
# calculate duration of each status for each equipment and each line in each hour for df_event
df_hour_list = []

for line in df_event['line'].unique():
    df_l = df_event[df_event['line'] == line]
    print(line)
    
    for equipment in df_l['equipment_name'].unique():
        df_e = df_l[df_l['equipment_name'] == equipment]

        print(equipment)
        # iterate over df_hour to calculate duration of each status for each equipment in each hour
        
        df_hour = pd.DataFrame()
        df_hour['datetime'] = pd.date_range(start_equipment, end_equipment, freq='H')
        
        df_hour['line'] = line
        df_hour['equipment_name'] = equipment
        
        for i, row in df_hour.iterrows():
            
            start = row['datetime']
            end = start + pd.to_timedelta(1, unit='h')
            
            for status in df_e['status'].unique():
                df_e_status = df_e[df_e['status'] == status]
                
                # calculate duration of each status for each equipment in each hour
                
                # Step 1: find status started this period, find status ended this period
                df_started_this_period = df_e_status[(df_e_status['start'] >= start) & (df_e_status['end'] < end)]
                # df_started_this_period_ended = df_started_this_period[df_started_this_period['end'] <= end]
                # sum_started_and_ended = df_started_this_period_ended['duration'].sum()
                sum_started_and_ended = df_started_this_period['duration'].sum()

                # Step 2: find status started before this period and ended after this period
                # df_started_this_period_not_ended = df_started_this_period[df_started_this_period['end'] > end]
                # sum_started_not_ended = (end - df_started_this_period_not_ended['start']).dt.total_seconds().sum()
                
                # Step 3: find status started before this period and ended in this period
                # df_started_before_this_period = df_e_status[(df_e_status['start'] < start) & (df_e_status['end'] >= start) & (df_e_status['end'] < end)]
                # sum_started_before_this_period = (df_started_before_this_period['end'] - start).dt.total_seconds().sum()
                
                # sum_all = sum_started_and_ended + sum_started_not_ended + sum_started_before_this_period
                sum_all = sum_started_and_ended
                
                df_hour.loc[i, '{}'.format(status)] = sum_all
                
        df_hour_list.append(df_hour)

```

```python
df_hour_all = pd.concat(df_hour_list)
df_hour_all
```

```python
# calculate total duration for each equipment
# check if data is correct
df_hour_all.groupby(['line', 'equipment_name'])['Error'].sum().reset_index()
```
