```python
# 获取订货单大小

# read from bom_batch_size.xlsx

def get_bom_batch_size():
    
    df = pd.read_excel(os.path.join(BASE_DIR, "bom_batch_size.xlsx"))
    
    return df
```

```python
df_size = get_bom_batch_size()
```

```python
df_size
```

```python
def get_min_batch_size_and_delivery(df_size, part_name):
    """return 3 values"""
    
    df_ = df_size[df_size['零件名称']==part_name]
    size_min = int(df_['size_min'].values[0])
    delivery_day = int(df_['delivery_day'].values[0])
    delivery_day_var = int(df_['delivery_day_var'].values[0])
    
    
    return size_min, delivery_day, delivery_day_var
    
    
```

```python
size_min, delivery_day, delivery_day_var = get_min_batch_size_and_delivery(df_size, '外壳')
```
