```python
def ge_real_delivery_day(delivery_day_var):
    """return real delivery day"""
    return int(generate_skewnorm(size = 1) * delivery_day_var)
```

```python
# 零部件的inv策略
# 1. 每月的开始，都会有一个月的库存，有最低库存量
# 2. 上个月的库存如果是负数，这个月就增加库存系数 

def part_inv_plan(inv_last_month, inv_min, inv_ratio = 1.1):
    """return inv min and inv max"""
    
    return max(inv_last_month * inv_ratio, inv_min)
```

```python
# 
INV_MIN_INIT = 5000
```

```python

def ge_po(dft, part_name = None, verbose = True):
    df_po = pd.DataFrame()

    
    need_part = 0
    inv_min = INV_MIN_INIT
    # 初始库存
    inv = inv_min
    inv_end = inv_min
    
    # iteration over month
    for idx, row in dft.iterrows():
        
        # inv 继承上一个周期的
        inv = inv_end
        
        need_this_month = row['count_raw']
        
        # 本周期需要的零部件数量
        need_part = need_this_month
        
        # 当前库存 - 本周期需要的零部件数量 应该大于 最小库存，否则需要订货
        if_need_to_order = 1 if inv - need_part <= inv_min else 0
        
        # 订货数量 = 需要的零部件数量 + 当前库存 - 最小库存
        need_order_size = need_part + inv - inv_min
        
        # 需要到货的日期
        should_arrive_date = row['month']
        # convert to datetime
        should_arrive_date = datetime.strptime(should_arrive_date, '%Y-%m-%d')
      
        # if need_part > size_min, then generate order
        if if_need_to_order:
            
            # generate order size, should be greater than size_min
            order_size = need_order_size if need_order_size > size_min else size_min
            
            # 应该下单时期 = 需要到货的日期 - 送货时间delivery_day
            should_order_date = should_arrive_date - timedelta(days=delivery_day)
            
            # generate delivery date accourding to delivery_day and delivery_day_var 
            # with skew gaussian distribution
            
            delivery_day_real = ge_real_delivery_day(delivery_day_var)
            
            # 实际到货时间 = 应该下单的日期 + 送货时间delivery_day + 送货时间波动delivery_day_var
            delivery_date_real = should_order_date + timedelta(delivery_day)+ timedelta(days=delivery_day_real)
            
            # print all information
            if not verbose:
              print('order_size: ', order_size, 'should_order_date: ', should_order_date, 'delivery_date_real: ', delivery_date_real)
              need_part -= order_size
              
              
            # 周期结束时，库存 = 库存 + 本周期需要的零部件数量 + 到货的零部件数量
            inv_end = inv - need_part + order_size
        
        # 不需要订货
        else:
            order_size = 0
            # should_arrive_date = None
            should_order_date = None
            delivery_date_real = None
            inv_end = inv - need_part
            # put order information in df_po
            
        df_po = df_po.append({
                                    'po_cycle': should_arrive_date,
                                    'inv_begin':inv,
                                    'inv_end':inv_end,
                                    'inv_min':inv_min,
                                    'need_order':if_need_to_order,
                                    'part_name': part_name,
                                    'need_this_month':need_this_month,
                                    'need_part':need_part,
                                    'order_size':order_size, 
                                  'should_order_date':should_order_date,
                                  'delivery_date_real':delivery_date_real}, 
                                ignore_index=True)
            
            # need_part minus order_size after this order
            
            
    return df_po
```
