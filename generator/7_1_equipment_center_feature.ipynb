```python
from imports import *
```

```python
# read equipment_status from GE dir

equipment_status = pd.read_csv(os.path.join(GE_DIR,'equipment_vibration.csv'))
```

```python
equipment_status = equipment_status.reset_index()
equipment_status
```

```python
rolling_std = equipment_status['noise_data'].rolling(window=1000, step=1000).std()
rolling_std = rolling_std.reset_index()
```

```python
rolling_std
```

```python
# rename columns in rolling_std
rolling_std.rename(columns={'noise_data':'rolling_std'}, inplace=True)
rolling_std
```

```python
rolling_std = rolling_std.reset_index()
```

```python
# time data
df = pd.merge(rolling_std, equipment_status, how='left', on ='index')
df
```

```python
px.scatter(df, x='index', y='rolling_std')
```

```python
# save to csv
df.to_csv(os.path.join(GE_DIR,'equipment_vibration_rolling_std.csv'), index=False)
```
