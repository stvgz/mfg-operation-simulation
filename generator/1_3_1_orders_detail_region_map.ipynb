# Generate map

```python
df_map = pd.read_json('../map_generation_src/map_china.json', orient='index')
```

```python
# geojson
```

```python
df_map
```

```python
obj = df_map.loc['features'][0]
d1 = pd.DataFrame.from_dict(obj)
d1
```

```python
d1['formated'] = d1.apply(lambda x: x.to_json(), axis=1)
d1['name_province'] = d1['properties'].apply(lambda x: x['name'])
```

# add sales

```python
df.head()
```

```python
df = df.rename(columns={'provice': 'name_province'})
```

```python
d1.head()
```

```python
d1_m = pd.merge(df, d1, on ='name_province', how='right')
```

```python
d1_m
```

```python
d1_m['count'] = d1_m['count'].fillna(0)
```

```python
d1_m.to_csv(os.path.join(GE_DIR, 'order_detail_map.csv'), index=False)
```
